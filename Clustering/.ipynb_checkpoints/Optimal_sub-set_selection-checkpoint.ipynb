# Functions

```python
def get_3D_tas_xarr(path, filename, variables):
#    “”"
#    This function read the netCDF file of monthly data, compute the radiative budget, perform a yearly mean and 
#    return a dataframe
#    “”"
    # First step : download the data into dataframe
    file = xr.open_mfdataset(path+filename,combine='by_coords')
    #
    # Second step : compute the annual average 
    df = file[variables].mean('time', keep_attrs=True)
    tas = df['tas']
    #
    return tas
```

```python
def MSE(mod, obs, W_rmse_2D) :
    diff_tmp = (mod - obs)**2 * W_rmse_2D
    diff = (diff_tmp.sum(['lat', 'lon']))
    return diff
```

```python
def reconstruct_X(eofs_combined, pc, nb_dims) :
    X_rec_tmp = np.dot(eofs_combined.transpose(),pc)
    if nb_dims == 3 :
        X_rec = xr.DataArray(X_rec_tmp, 
                        dims=["lon", "lat", "time"]).transpose('time', 'lat', 'lon')
    if nb_dims == 2 :
        X_rec = xr.DataArray(X_rec_tmp, 
                        dims=["lon", "lat"]).transpose('lat', 'lon')
    
    X_rec['lat'] = eofs_combined['lat']
    X_rec['lon'] = eofs_combined['lon']
    return X_rec
```

```python
def MSE_rec(rec_anom_mod_w, rec_anom_obs_w, Mean, W_rmse_2D) :
    mod = rec_anom_mod_w/W_eof_2D + Mean
    obs = rec_anom_obs_w/W_eof_2D + Mean
    diff = MSE(mod, obs, W_rmse_2D)
    return diff
```

# Test de sensibilité au nb de cluster

```python
nb_mbs_PPE = {}
nb_mbs_pred = {}

for n_clusters in range(3,10,1) :
    print('n_cluster = '+str(n_clusters))
    kmean = KMeans(n_clusters).fit(X)
    cluster = kmean.labels_
    df_X['cluster'] = cluster

    for c in range(0,n_clusters,1) :
        print('c = '+str(c))
        tmp = df_X.groupby('cluster').get_group(c)
        nb_mbs_PPE['n_cluster = '+str(n_clusters)+', c = '+str(c)] = tmp.groupby('legend').get_group('PPE')
        nb_mbs_pred['n_cluster = '+str(n_clusters)+', c = '+str(c)] = tmp.groupby('legend').get_group('pred')
```

```python
df_mbs_PPE = {}
df_mbs_pred = {}
for n_clusters in range(3,10,1) :
    mbs_PPE = []
    mbs_pred = []
    for c in range(0,n_clusters,1) :
        tmp = len(nb_mbs_PPE['n_cluster = '+str(n_clusters)+', c = '+str(c)])
        mbs_PPE.append(tmp)
        tmp = len(nb_mbs_pred['n_cluster = '+str(n_clusters)+', c = '+str(c)])
        mbs_pred.append(tmp)
    df_mbs_PPE['n_cluster = '+str(n_clusters)] = mbs_PPE
    df_mbs_pred['n_cluster = '+str(n_clusters)] = mbs_pred
```

```python
df_fig = {}

for c in range(0,10,1) :
    for n_clusters in range(3,10,1) :
        tmp_list = []
        tmp = df_mbs_PPE['n_cluster = '+str(n_clusters)][c]
        tmp_list.append(tmp)
    df_fig['cluster '+str(c)] = tmp_list
```
